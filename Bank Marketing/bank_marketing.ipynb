```python
# okay so this dataset has a lot of strings. we want numbers.
# so I'll label encode stuff depending on the number of unique values.

strings = ["job", 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']

df[strings].nunique()
```

```python
df.poutcome.value_counts()
```

```python
def month_encoder(month):
    dict = {'jan': 1, 'feb': 2, 'mar': 3, 'apr' : 4, 'may' : 5, 'jun': 6, 
            'jul': 7, 'aug': 8, 'sep': 9, 'oct' : 10, 'nov' : 11, 'dec' : 12}
    return dict[month]

df['month'] = df['month'].apply(month_encoder)
```

```python
df.head()
```

```python
df.info()
```

```python
df.contact.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

cols = ["job", 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome', 'y']

for col in cols:
    df[col] = label.fit_transform(df[col])
```

```python
df.head()
```

```python
# classification problem.
# let's try the usual algos

X = df.drop('y', axis = 1)
y = df['y']
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log = LogisticRegression()
log.fit(X_train, y_train)
Prediction = log.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, Prediction)))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.naive_bayes import GaussianNB

gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
y_pred = gaussian.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(random_state=1)
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes = (5, 2), random_state = 1)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
```
